```python
def check_input(element):
    return isinstance(element, str)


def get_substring(main_str):
    left = 0
    char_set = set()
    max_len = 0
    longest_str = ""
    for i in range(len(main_str)):
        while main_str[i] in char_set:
            char_set.remove(main_str[left])
            left += 1
        char_set.add(main_str[i])
        new_line = i - left + 1
        if new_line > max_len:
            max_len = new_line
            longest_str = main_str[left:i + 1]
    return longest_str

def main(user_input):
    str_only = []
    for i, item in enumerate(user_input):
        if check_input(item):
          str_only.append((i, item))
    longest_str_info = max(str_only, key = lambda x: len(x[1]))
    longest_str_index = longest_str_info[0]
    longest_str_value = longest_str_info[1]
    substring = get_substring(longest_str_value)
    print(f"{longest_str_index}, '{substring}'")

user_input = [[0, 4, 8], "abca", 7, "aaaaa"]
main(user_input)
```
